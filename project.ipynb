
Python Restaurant Ordering System

This is short, clear, and tells exactly what the program does.


```python
menu = {
    'water bottle':20,
    'chai':15,
    'coffee':30,
    'thali':80,
    'chocolate':25
}

#greet
print("welcome to python resturant")
print("Water bottle :rs20\nChai :rs15\nCoffee: rs30\nThali: rs80\nChocolate: rs25")
order_total = 0
#30 + 20 = 50

item_1 = input("enter the name of item").lower()
if item_1 in menu:
    order_total += menu [item_1]
    print(f"your item {item_1} has been added to your order")

else:
    print(f"ordered item{item_1} is not avaialable yet")

another_order = input("do you want to add another item? (yes/no)")
if another_order== "yes":
    item_2 = input("enter the name of second item =")
    if item_2 in menu:
        order_total+=menu[item_2]
        print(f"item{item_2} has been added to order")
    else:
        print(f"ordered item {item_2}is not avaialable")

print(f"the total amount of items to pay is {order_total}")
print(f"Total Amount to Pay: Rs {order_total}")
print(f"Thank you for dining with Python Restaurant")

    
```
